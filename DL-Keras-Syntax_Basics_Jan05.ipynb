**Scaling the Data**

```python
from sklearn.preprocessing import MinMaxScaler  # scaling'den con normalizasyona benziyor
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

**TensorFlow Syntax**

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Activation
```

```python
model = Sequential()
model.add(Dense(4, activation = 'relu'))     # input_dim = 2, input_dim = X_train.shape[0];   4 hidden layer neuron sayisi
model.add(Dense(4, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer = 'rmsprop', loss = 'mse')
```

```python
model.fit(X_train, y_train, batch_size = 32, epochs = 250)
```

```python
model.summary()
```

Tablodaki param sayisinin acilimi:

```python
model2 = Sequential()
model2.add(Dense(4, input_dim = 2, activation = 'relu'))     # input_dim = 2, input_dim = X_train.shape[0];   4 hidden layer neuron sayisi
model2.add(Dense(4, activation = 'relu'))
model2.add(Dense(4, activation = 'relu'))
model2.add(Dense(1))
```

```python
model2.compile(optimizer = 'rmsprop', loss = 'mse')
```

```python
model2.fit(X_train, y_train, batch_size = 16, epochs = 250)
```

```python
model2.summary()
```

**Deep Learning outlier'lara sensitive'dir - scale etmek genelde gereklidir.**

```python
loss_df=pd.DataFrame(model2.history.history)
```

```python
loss_df.plot()
```

```python
loss = model2.history.history['loss']
sns.lineplot(x=range(len(loss)),y=loss)
plt.title("Training Loss per Epoch");
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\n","mae:", mae, "\n","mse:",mse, "\n","rmse:",rmse)
```

```python
y_pred = model2.predict(X_test)
```

```python
eval_metrics(y_test, y_pred)
```
